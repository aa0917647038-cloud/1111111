## 第一步：要設定兩個資訊：

壓縮檔叫什麼名字？ (通常是「XX.zip 的副本」)

請在下方的程式碼中，修改引號裡面的文字。

```python
import os
import zipfile

# --- 這裡請根據你的檔案名稱修改 ---
# 1. 你從老師那邊複製過來的壓縮檔檔名 (請包含 .zip)
target_zip_name = "M4-CNN.zip 的副本"

# 2. 你想要建立的新資料夾名稱 (例如：01_環境介紹)
new_folder_name = "M4-CNN"
# ------------------------------

# 指定你要放資料的雲端路徑
base_path = "/content/drive/MyDrive/"
base_path2 = "/content/drive/MyDrive/_202604南大AI"

# 原路徑
zip_file_path = os.path.join(base_path, target_zip_name)

# 解壓的新路徑
extract_path = os.path.join(base_path2, new_folder_name)

print(f"準備處理檔案：{target_zip_name}")
```

## 第二步：自動建立資料夾並「開箱」
最後這段程式碼會幫我們檢查：

資料夾是否存在？如果不存在，它會自動幫你「買一個新櫃子」（建立資料夾）。

使用工具把 .zip 裡面的所有檔案「搬」進去。

```python
# 檢查新資料夾是否已經存在
if not os.path.exists(extract_path):
    # 如果不存在，就建立一個新的資料夾
    os.makedirs(extract_path)
    print(f"📂 已經為您建立新資料夾：{new_folder_name}")
else:
    print(f"ℹ️ 資料夾 '{new_folder_name}' 已經存在，準備覆蓋或存入新檔案。")

# 開始解壓縮
try:
    # 打開那個壓縮檔
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        # 將裡面的內容全部解壓到剛剛設定的新資料夾
        zip_ref.extractall(extract_path)

    print("---")
    print(f"🎯 太棒了！解壓縮成功！")
    print(f"📍 你的檔案現在躺在：雲端硬碟 > {new_folder_name} 資料夾裡囉。")

except FileNotFoundError:
    # 如果找不到檔案，會顯示這段提醒
    print("❌ 錯誤：找不到壓縮檔！請檢查你的檔案名稱是否完全正確（包含空格與副本二字）。")
except Exception as e:
    # 如果發生其他意外，會顯示錯誤訊息
    print(f"❌ 發生了點小問題：{e}")
```
